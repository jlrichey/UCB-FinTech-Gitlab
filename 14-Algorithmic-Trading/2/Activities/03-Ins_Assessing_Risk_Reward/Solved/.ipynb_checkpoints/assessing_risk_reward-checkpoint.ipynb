# Assessing the Risk/Reward Characteristics of a Trading Algorithm

## Load Data

```python
# Import the required libraries and dependencies
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
# Read the trading_signals.csv file into a Pandas DataFrame
# Set the date column as the DateTimeIndex
signals_df = pd.read_csv(
    Path("../Resources/trading_signals.csv"),
    index_col="date",
    parse_dates=True,
    infer_datetime_format=True)

# Review the DataFrame
signals_df.tail(10)
```

## Set Up the Portfolio-Level Risk/Reward Evaluation Metrics

```python
# Create a list for the column name
columns = ["Backtest"]

# Create a list holding the names of the new evaluation metrics
metrics = [
    "Annualized Return",
    "Cumulative Returns",
    "Annual Volatility",
    "Sharpe Ratio",
    "Sortino Ratio"]

# Initialize the DataFrame with index set to the evaluation metrics and the column
portfolio_evaluation_df = pd.DataFrame(index=metrics, columns=columns)

# Review the DataFrame
portfolio_evaluation_df
```

## Calculate the Metrics

### Annualized Return

```python
# Calculate annualized return
portfolio_evaluation_df.loc["Annualized Return"] = (
    signals_df["Portfolio Daily Returns"].mean() * 252
)

# Review the result
portfolio_evaluation_df
```

### Cumulative Returns

```python
# Calculate cumulative return
portfolio_evaluation_df.loc["Cumulative Returns"] = signals_df["Portfolio Cumulative Returns"][-1]

# Review the result
portfolio_evaluation_df
```

```python
# Calculate cumulative return
cumulative_return = signals_df["Portfolio Cumulative Returns"].iloc[-1]

# Add the cumulative return to the evaluation DataFrame
portfolio_evaluation_df.loc["Cumulative Returns"] = cumulative_return

# Review the result
portfolio_evaluation_df
```

### Annual Volatility

```python
# Calculate annual volatility
portfolio_evaluation_df.loc["Annual Volatility"] = (
    signals_df["Portfolio Daily Returns"].std() * np.sqrt(252)
)

# Review the result
portfolio_evaluation_df
```

### Sharpe Ratio

```python
# Calculate Sharpe ratio
portfolio_evaluation_df.loc["Sharpe Ratio"] = (
    signals_df["Portfolio Daily Returns"].mean() * 252) / (
    signals_df["Portfolio Daily Returns"].std() * np.sqrt(252)
)

# Review the result
portfolio_evaluation_df
```

### Sortino Ratio

#### Calculate downside return values

```python
# Create a DataFrame that contains the Portfolio Daily Returns column
sortino_ratio_df = signals_df[["Portfolio Daily Returns"]].copy()

# Create a column to hold downside return values
sortino_ratio_df.loc[:,"Downside Returns"] = 0

# Find Portfolio Daily Returns values less than 0,
# square those values, and add them to the Downside Returns column
sortino_ratio_df.loc[sortino_ratio_df["Portfolio Daily Returns"] < 0,
                     "Downside Returns"] = (sortino_ratio_df["Portfolio Daily Returns"]**2).astype('int64')

# Review the DataFrame
sortino_ratio_df.tail()
```

```python
# Replace non-finite values with a placeholder value (e.g., 0)
sortino_ratio_df["Portfolio Daily Returns"].fillna(0, inplace=True)

# Convert the "Downside Returns" column to int64 before assignment
sortino_ratio_df.loc[:, "Downside Returns"] = sortino_ratio_df["Downside Returns"].astype('int64')

# Find Portfolio Daily Returns values less than 0,
# square those values, and add them to the Downside Returns column
sortino_ratio_df.loc[sortino_ratio_df["Portfolio Daily Returns"] < 0,
                     "Downside Returns"] = (sortino_ratio_df["Portfolio Daily Returns"]**2).astype('int64')
```

#### Calculate the Sortino ratio

```python
# Calculate the annualized return value
annualized_return = (
    sortino_ratio_df["Portfolio Daily Returns"].mean() * 252
)

# Print the result
print(f"Annualized Return: {annualized_return}")
```

```python
# Calculate the annualized downside standard deviation value
downside_standard_deviation = (
    np.sqrt(sortino_ratio_df["Downside Returns"].mean()) * np.sqrt(252)
)

# Print the result
print(f"Annualized Downside STD: {downside_standard_deviation}")
```

```python
# The Sortino ratio is reached by dividing the annualized return value
# by the downside standard deviation value
sortino_ratio = annualized_return/downside_standard_deviation

# Add the Sortino ratio to the evaluation DataFrame
portfolio_evaluation_df.loc["Sortino Ratio"] = sortino_ratio

# Review the DataFrame
portfolio_evaluation_df
```

```python
# Check if downside_standard_deviation is not zero to avoid divide by zero error
if downside_standard_deviation != 0:
    # Calculate the Sortino ratio
    sortino_ratio = annualized_return / downside_standard_deviation
else:
    # Handle the case where downside_standard_deviation is zero
    # For example, set the sortino_ratio to zero or raise an error
    sortino_ratio = 0  # You can replace 0 with any predefined value or handle it according to your requirements

# Add the Sortino ratio to the evaluation DataFrame
portfolio_evaluation_df.loc["Sortino Ratio"] = sortino_ratio

# Review the DataFrame
portfolio_evaluation_df
```

## Set Up the Trade-Level Risk/Reward Evaluation Metrics

```python
# Loop through signal DataFrame
# If `Entry/Exit` is 1, set entry trade metrics
# Else if `Entry/Exit` is -1, set exit trade metrics and calculate profit
# Then append the record to the trade evaluation DataFrame

# Initialize an empty list to hold the trade data
trades_list = []

for index, row in signals_df.iterrows():
    if row["Entry/Exit"] == 1:
        entry_date = index
        entry_portfolio_holding = row["Portfolio Holdings"]
        share_size = row["Entry/Exit Position"]
        entry_share_price = row["close"]

    elif row["Entry/Exit"] == -1:
        exit_date = index
        exit_portfolio_holding = abs(row["close"] * row["Entry/Exit Position"])
        exit_share_price = row["close"]
        profit_loss = exit_portfolio_holding - entry_portfolio_holding

        # Instead of appending to DataFrame, append the record as a dict to the list
        trades_list.append({
            "Stock": "AAPL",
            "Entry Date": entry_date,
            "Exit Date": exit_date,
            "Shares": share_size,
            "Entry Share Price": entry_share_price,
            "Exit Share Price": exit_share_price,
            "Entry Portfolio Holding": entry_portfolio_holding,
            "Exit Portfolio Holding": exit_portfolio_holding,
            "Profit/Loss": profit_loss
        })

# After the loop, create a DataFrame from the list of dictionaries
trade_evaluation_df = pd.DataFrame(trades_list)

# Print the DataFrame
print(trade_evaluation_df)
```
