# K-Means Clustering

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import pandas as pd
import hvplot.pandas
```

## Importing the dataset

```python
dataset = pd.read_csv(Path('../Resources/Mall_Customers.csv'))
display(dataset)
```

```python
dataset.hvplot.scatter(x="Spending Score (1-100)", y="Annual Income (k$)") #, by="Genre"
```

## Run the k-means model with 2 clusters

```python
# Start by importing the K-means algorithm
from sklearn.cluster import KMeans
```

```python
# Select your columns that you are trying to understand
X = dataset.iloc[:, [3, 4]].values
X
```

```python

# Create and initialize the K-means model instance for 2 clusters
model = KMeans(n_clusters=2, random_state=1)

# Print the model
model
```

```python
# Fit the data to the instance of the model
model.fit(X)
```

```python
# Make predictions about the data clusters using the trained model
clusters = model.predict(X)

# Print the predictions
print(clusters)
```

```python
# Create a copy of the DataFrame
dataset_predictions_df = dataset.copy()

# Add a column to the DataFrame that contains the customer_ratings information
dataset_predictions_df['clusters_prediction'] = clusters

# Review the DataFrame
dataset_predictions_df.head()
```

```python
# Plot the data points based on the customer rating
dataset_predictions_df.hvplot.scatter(x="Annual Income (k$)", y="Spending Score (1-100)" , by="clusters_prediction")
```

## Using the elbow method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans
wcss = [] # The squared distance between each point and the centroid in a cluster
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
wcss.sort()
wcss
```

## Training the K-Means model on the dataset

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

## Visualising the clusters

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
