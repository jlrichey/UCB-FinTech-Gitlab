```python
# Import dependencies
import pandas as pd
from pathlib import Path
import hvplot.pandas
```

```python
# Read in the CSV file as a Pandas DataFrame
service_ratings_df = pd.read_csv(
    Path("../Resources/service_ratings.csv")
)

# Review the DataFrame
service_ratings_df.head()
```

```python
# Scatter plot of data
service_ratings_df.hvplot.scatter(
    x="mobile_app_rating", 
    y="personal_banker_rating"
)
```

## Run the k-means model with 2 clusters

```python
# Start by importing the K-means algorithm

from sklearn.cluster import KMeans
```

```python
# Create and initialize the K-means model instance for 2 clusters
model = KMeans(n_clusters=2, random_state=1)

# Print the model
model
```

```python
# Fit the data to the instance of the model
model.fit(service_ratings_df)
```

```python
# Make predictions about the data clusters using the trained model

customer_ratings = model.predict(service_ratings_df)
# Print the predictions
customer_ratings
```

```python
# Create a copy of the DataFrame
customer_ratings_predictions_df = service_ratings_df.copy()

# Add a column to the DataFrame that contains the customer_ratings information
customer_ratings_predictions_df['cluster_prediction'] = customer_ratings

# Review the DataFrame
customer_ratings_predictions_df
```

```python
# Plot the datapoints based on the customer rating
customer_ratings_predictions_df.hvplot.scatter(x="mobile_app_rating", y="personal_banker_rating", by="cluster_prediction")
```
